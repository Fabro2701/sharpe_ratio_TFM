```python
from core.dgp import ARProcess, ARGARCHProcess, IIDProcess, NormalInnov, StudentTInnov
from core.synth import TrajectorySpec, SyntheticGenerator, summary

```

```python
specs = [TrajectorySpec(IIDProcess(StudentTInnov(df=2.5).calibrate_params(mu=1.5, sigma=1)),    
                        "iid_student",  n=1000, length=10000)]

data = SyntheticGenerator(seed=42).generate(specs)
print(data)
print(summary(data))
```

```python
specs = [TrajectorySpec(ARProcess(phi=[0.8], drift=2, innov=StudentTInnov()).calibrate_params(mu=1.5, sigma=0.4),    
                        "ar1_student",  n=100, length=10000),
        TrajectorySpec(ARProcess(phi=[0.8], drift=2, innov=NormalInnov()).calibrate_params(mu=1.5, sigma=0.4),    
                        "ar2_normal",  n=100, length=10000)]

data = SyntheticGenerator(seed=42).generate(specs)
print(data)
print(summary(data))
```

```python
specs = [TrajectorySpec(ARProcess(phi=[0.8], drift=2, innov=NormalInnov()).calibrate_params(mu=1.5, sigma=0.4),    
                        "ar2_normal",  n=1, length=10000)]

data = SyntheticGenerator(seed=42).generate(specs)
print(summary(data))
```

```python
from core.models import AR1NormalModel, IIDStudentTModel, AR1NonNormalModel
```

```python
model = AR1NormalModel()
model.fit(data.loc[('ar2_normal', 0)].value.values)
```

```python
specs = [TrajectorySpec(IIDProcess(StudentTInnov(df=5.5).calibrate_params(mu=1.5, sigma=1)),    
                        "iid_student",  n=1, length=10000)]

data = SyntheticGenerator(seed=42).generate(specs)
print(summary(data))

model = IIDStudentTModel()
print(model.fit(data.loc[('iid_student', 0)].value.values))
```
